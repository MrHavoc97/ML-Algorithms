```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
X_train = college_data.drop(['Private', 'Unnamed: 0'], axis=1)
```

```python

kmeans.fit(X_train)
```

```python
kmeans.cluster_centers_
```

```python
college_data['Cluster'] = college_data['Private'].apply(lambda cluster: 1 if cluster == 'Yes' else 0)
```

```python
college_data.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
sns.heatmap(confusion_matrix(college_data['Cluster'],kmeans.labels_), annot=True)
```

```python
print(classification_report(college_data['Cluster'],kmeans.labels_))
```
