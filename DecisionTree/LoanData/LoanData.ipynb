```python
import pandas as pd 
import numpy as np 

import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
loan = pd.read_csv('loan_data.csv')
loan.head()
```

```python
loan.info()
```

```python
loan.describe()
```

```python
plt.figure(figsize=(10,6))
loan[loan['not.fully.paid']==1]['fico'].hist(alpha=0.5, color='blue', bins=30,label='not.fully.paid=1')
loan[loan['not.fully.paid']==0]['fico'].hist(alpha=0.5,color='red', bins=30,label='not.fully.paid=0')
plt.legend()
plt.xlabel('FICO')
```

```python
sns.jointplot(x='fico',y='int.rate',data=loan, palette='viridis')
```

```python
plt.figure(figsize=(11,7))
sns.lmplot(y='int.rate',x='fico',data=loan, hue='credit.policy', col='not.fully.paid',palette='viridis')
```

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loan, columns=cat_feats, drop_first=True)
```

```python
final_data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=45)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
sns.heatmap(confusion_matrix(y_test,predictions), annot=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
sns.heatmap(confusion_matrix(y_test,predictions), annot=True) 
```
